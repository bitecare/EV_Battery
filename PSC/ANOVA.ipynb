```python
#rta1

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols


# RTA1 공정의 변수
rta1_cols = ['rta1_box_col', 'rta1_box_row', 'rta1_box_dan']

# RTA1 공정에 대해 ANOVA 분석 수행
formula = 'defect_rate ~ ' + ' + '.join(rta1_cols)
model = ols(formula, data=df_tray).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# ANOVA 결과 출력
print("ANOVA results for RTA1 process:")
print(anova_table)
```
