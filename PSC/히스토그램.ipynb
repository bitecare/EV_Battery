```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../OUT/data_fin.csv")
```

```python
good = df[df["judge"] == "양품"]
bad = df[df["judge"] == "불량"]
```

```python
plt.figure()
for judge, color in zip(['양품', '불량'], ['blue', 'red']):
    sns.kdeplot(df[df['judge'] == judge]['c1_temp_avg'], label=judge, color=color, shade=False, linewidth=2)

# 그래프 제목과 레이블 설정
plt.title("Density Plot by Judge")
plt.xlabel("c1_temp_avg")
plt.ylabel("Density")
plt.legend(title='Judge')

# 그래프 표시
plt.show()
```

```python
plt.figure()
for judge, color in zip(['양품', '불량'], ['blue', 'red']):
    sns.kdeplot(df[df['judge'] == judge]['dc1_capafit'], label=judge, color=color, shade=False, linewidth=2)

# 그래프 제목과 레이블 설정
plt.title("Density Plot by Judge")
plt.xlabel("c1_temp_avg")
plt.ylabel("Density")
plt.legend(title='Judge')

# 그래프 표시
plt.show()
```

```python
df["c1_temp_avg"] = df["c1_temp_avg"]/10

interval1 = df[df["c1_temp_avg"] < 27]
interval2 = df[(df["c1_temp_avg"] >= 27) & (df["c1_temp_avg"] <= 29)]
interval3 = df[df["c1_temp_avg"] > 29]
lists = [interval1, interval2, interval3]
cnt = 1
for i in lists:
    print(f"interval{cnt} count: {i.judge.value_counts()}")
    print(" ")
    print(f"interval{cnt} proportion : {i.judge.value_counts() / len(i)}")
    print(" ")
    cnt += 1
```
